# Importing necessary modules

```python
import os
import pandas as pd
```

```python
os.listdir("./Dataset/raw_text")
```

# Joining all the partitions text files in a single csv file

```python
data_dir = './Dataset/raw_text'                       # Text data files should be in this directory
train_file = os.path.join(data_dir, 'train.txt')      # Train part file
val_file = os.path.join(data_dir, 'val.txt')          # Val part file
test_file = os.path.join(data_dir, 'test.txt')        # Test part file
challenge_file = os.path.join(data_dir, 'challenge_test_set.txt')       # Challenge Part file

df_train = pd.read_csv(train_file, sep = '\t', header = None)           # Load Train file
df_train['DTYPE'] = 'train' 
df_val = pd.read_csv(val_file, sep = '\t', header = None)               # Load Val file
df_val['DTYPE'] = 'val' 
df_test = pd.read_csv(test_file, sep = '\t', header = None)             # Load Test file
df_test['DTYPE'] = 'test' 
df_challenge = pd.read_csv(challenge_file, sep = '\t', header = None)   # Load Challenge file
df_challenge['DTYPE'] = 'challenge' 

df = pd.concat([df_train, df_val, df_test, df_challenge], ignore_index = True)    # Concatenate all partitions
df.columns = ['image_id','X','Y','Width','Height','English Text','Hindi Text', 'P_TYPE']
df
```

# Saving final DataFrame

```python
df.to_csv(os.path.join('./Dataset/text',"HVG.csv"), index = None)
df
```
