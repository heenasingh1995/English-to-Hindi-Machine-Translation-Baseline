# Importing necessary modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Text Only

### Reading Dataset

```python
typ = 'text_only'
df1 = pd.read_csv("2_Train_metrics_30_05_2023_17_17_28.csv")
df1['epoch'] = list(range(1, len(df1)+1))
df1['Model'] = 'Text Only'
```

### Epoch vs BLEU Score

```python
fig, ax = plt.subplots(figsize=(7, 6))
ax.set_axisbelow(True)
max_epoch = df1['bleu_main'].argmax()
ax.plot(df1['epoch'], df1['bleu_main'],'darkred')
ax.plot(df1['epoch'].iloc[max_epoch], df1['bleu_main'].iloc[max_epoch],'o',color = 'darkblue')
s = "Best BLEU Score : {} \nEpoch : {}".format(df1['bleu_main'].iloc[max_epoch], df1['epoch'].iloc[max_epoch])
ax.text(121.5, 39.5, s, fontsize=9, horizontalalignment='left',verticalalignment='center',bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
ax.annotate('', xy=(df1['epoch'].iloc[max_epoch], df1['bleu_main'].iloc[max_epoch]), xytext=(120, 37.7), arrowprops=dict(
               width=2, headwidth=8, facecolor='wheat', shrink=0.001)) 
ax.spines[['right', 'top']].set_visible(False)
ax.grid(color='dimgrey', linestyle='-', linewidth=0.1)
plt.xticks(np.arange(0, 230, step=10), rotation = 60, color ='black')
plt.yticks(np.arange(0, 45, step=2), color =(0,0,0))
#plt.title("Text only Mod")
plt.xlabel("Epoch")
plt.ylabel("BLEU Score")
plt.savefig(typ+"_epoch_bleu.png",dpi = 300)
plt.show()
```

### Loss Graphs

```python
fig, ax = plt.subplots(figsize=(7, 6))
ax.set_axisbelow(True)
ax.plot(df1['epoch'], df1['Train_Loss'],'darkred', label='Training Loss')
ax.plot(df1['epoch'], df1['Val_Loss'],'green', label='Validation Loss')
ax.spines[['right', 'top']].set_visible(False)
plt.grid(color='dimgrey', linestyle='-', linewidth=0.1)
plt.xticks(np.arange(0, 230, step=10), rotation = 60, color ='black')
plt.yticks(np.arange(0, 6, step=0.3), color =(0,0,0))
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(facecolor = 'wheat')
plt.savefig(typ+"_epoch_loss.png",dpi = 300)
plt.show()
```

# Speech

### Reading Data

```python
df2 = pd.read_csv("3_Train_metrics_30_05_2023_17_06_21.csv")
df2['epoch'] = list(range(1, len(df2)+1))
df2['Model'] = 'Text with Speech'
df2
```

### Epochs vs BLEU

```python
typ = 'speech'
fig, ax = plt.subplots(figsize=(7, 6))

ax.set_axisbelow(True)
max_epoch = df2['bleu_main'].argmax()
ax.plot(df2['epoch'], df2['bleu_main'],'darkred')
ax.plot(df2['epoch'].iloc[max_epoch], df2['bleu_main'].iloc[max_epoch],'o',color = 'darkblue')
s = "Best BLEU Score : {} \nEpoch : {}".format(df2['bleu_main'].iloc[max_epoch], df2['epoch'].iloc[max_epoch])
plt.text(279, 41.7, s, fontsize=9, horizontalalignment='left',verticalalignment='center',bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
plt.annotate('', xy=(df2['epoch'].iloc[max_epoch], df2['bleu_main'].iloc[max_epoch]), xytext=(492, 40), arrowprops=dict(
               width=2, headwidth=8, facecolor='wheat', shrink=0.001)) 
ax.spines[['right', 'top']].set_visible(False)
plt.grid(color='dimgrey', linestyle='-', linewidth=0.1)
plt.xticks(np.arange(0, 720, step=30), rotation = 60, color ='black')
plt.yticks(np.arange(0, 47, step=2), color =(0,0,0))
#plt.title("Text only Mod")
plt.xlabel("Epoch")
plt.ylabel("BLEU Score")
plt.savefig(typ+"_epoch_bleu.png",dpi = 300)
plt.show()
```

### Loss Graphs 

```python
fig, ax = plt.subplots(figsize=(7,6))
ax.set_axisbelow(True)
ax.plot(df2['epoch'], df2['Train_Loss'],'darkred', label='Training Loss')
ax.plot(df2['epoch'], df2['Val_Loss'],'green', label='Validation Loss')
ax.spines[['right', 'top']].set_visible(False)
plt.grid(color='dimgrey', linestyle='-', linewidth=0.1)
plt.xticks(np.arange(0, 720, step=30), rotation = 60, color ='black')
plt.yticks(np.arange(0, 6, step=0.3), color =(0,0,0))
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(facecolor = 'wheat')
plt.savefig(typ+"_epoch_loss.png",dpi = 300)
plt.show()
```
